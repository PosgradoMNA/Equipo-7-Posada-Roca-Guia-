# **Parte 1: Fundamentos de bases de datos**
##1. Fundamentos de bases de datos y para ciencia de datos.  

La ciencia de datos es uno de los campos de estudio con tendencia en el sector empresarial. Debido a que la tendencia de la toma de decisiones con base en los datos se va haciendo cada vez más grande, por ello se requiere de personas capacitadas para el conocimiento de los datos, obtenerlos, limpiarlos, analizarlos, visualizarlos y usarlos para construir soluciones. Para esto, el rol del científico de datos es fundamental para crear las bases de datos y su debido tratamiento con soluciones y metodologías de científicos de datos requeridos en los proyectos específicos.

Estas actividades consisten principalmente en recolectar y analizar los datos de diversas fuentes para su posterior análisis y comprensión a través de la aplicación de múltiples herramientas, para hacer estructuras sencillas, pero, a su vez eficientes, seguras y rápidas. Una base datos está definida como una estructura de conjuntos de datos almacenados en una memoria de la computadora o en la nube, y pueden tener entidades, atributos y relaciones de la información de una organización.
 Estos datos provienen de fuentes de información variadas y pueden ser almacenadas de distintas formas conteniendo como elemento general registros y campos, definidas como tablas. Otra forma de verlo es como una colección de datos relacionados a las tablas donde dichas tablas tienen relaciones entre los datos que van obteniendo, esto definiendo la estructura central de las bases de datos.

Para las empresas, se busca que las bases de datos es básicamente, tener repositorios centrales, datos gestionados por un agente y almacén de forma estadística y conveniente al usuario. Para ello usan sistemas DBMS (Database Management Systems). Esto se define como un software que se utiliza para crear, mantener y proporcionar acceso controlado a los usuarios de una base de datos. Este programa interactúa con la base de datos mediante una solicitud ad hoc al DBMS. Los dos lenguajes más conocidos:
1. DML, también conocido como lenguaje de consulta (Query)
2. SQL (Structured Query Language), siendo este el más utilizado.

Con SQL viene una herramienta poderosa para manipular los datos. Ya que es fácil de usar, poderosa y eficiente. Gracias a este lenguaje las bases de datos se han dividido en dos tipos:

1. Bases de datos relacionables: estas son las que los datos están organizados y almacenados dentro de tablas que pueden ser conectadas a otras con alguna clase de relación.
2. Bases de datos no relacionables: Estas bases de datos también son llamadas NoSQL. Debido a que estas consiguen conectarse por medio de categorías más que relaciones entre las mismas.


###2. Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.
Un almacén de datos es básicamente una base de datos gigante que contiene información de diferentes fuentes y las ordena o clasifica para una determina función. Estas comúnmente están diseñadas en una relación normalizada de una base de datos, ya que tienden a contener datos históricos de accesos de datos del pasado con los actuales, tablas actualizadas y versiones pasadas que muchas veces son actualizadas en tiempo real, etc.

Las técnicas de almacenamiento de datos en repositorios centrales de información facilitan las técnicas de análisis para la tomar decisiones mejor informadas, para esto, los datos fluyen hacia estos almacenes desde sistemas transaccionales, bases de datos relacionales y otros orígenes, normalmente a una cadencia regular. Dentro de ello logra centralizar y consolidar toda esta información masiva llamada también Big Data. Dotando al usuario una manera de ejecutar métodos de análisis de manera que tenga este contenedor central que lo ayude y por ende extraer información valiosa de los mismos.

La arquitectura de las bases de datos se basa en:

1. Simple: Todos los almacenes de datos comparten un diseño simple en donde la metadata, el resumen de los datos y el bruto de la data están almacenados dentro del repositorio central del almacén.
2. Simple con escenificación del área: Los datos operacionales deben ser limpiados y procesados antes de ser puestos en el almacén. Aunque esto puede ser ejecutado por algún programa, muchos almacenes de dados añaden un área de puesta de escena para los datos, antes que estos entren al almacén.
3. Centro y habla: Añadir mercados de datos entre los repositorios centrales y los usuarios finales permiten a la organización personalizar sus almacenes de datos para servir a las diferentes líneas del negocio.
4. Cajas de arena: Las cajas de arena son privadas y seguras. Estas áreas seguras permiten a las compañías explorar e informar rápidamente acerca de nuevos conjuntos de datos o maneras de analizar información sin tener que conformarse a solo compilarla dentro de las reglas y protocolos tradicionales de los almacenes de datos.

Dentro de los beneficios de un almacenamiento de datos se incluyen los siguientes:

- Toma de decisiones fundamentadas
- Datos consolidados provenientes de muchos orígenes
- Análisis de datos históricos
- Calidad, coherencia y precisión de los datos
- Separación del procesamiento de análisis de las bases de datos transaccionales, lo que mejora el rendimiento de ambos sistemas

###**Método de limpieza 2**

```python
#PAra este segundo método de limpieza, tambien iniciamos copiando el dataset previo, para evitar pérdidas de información
ndf_MeanWh = df2.copy()
#Verificamos el contenido del DataSet
ndf_MeanWh
```

```python
#El siguiente paso consiste en extraer los promedios de cada columna, para ello, se utilizará un ciclo for, que recorra todas las filas.
name = ndf_MeanWh.columns
print(name)
Mean = []
for n in name:
  Mean.append(ndf_MeanWh[n].mean()) #using Append to save these values in a list
print(Mean)
```

```python
#En esta parte, se sustituiran todos los elementos vacios del data set con valores promedio generales.
name = ndf_MeanWh.columns

for n in range(0,len(name)):
  j=name[n]
  ndf_MeanWh[j].fillna(value = Mean[n], inplace = True)
  
ndf_MeanWh
```

```python
#Aplicamos la función mean, para obtener la media de los datos recien reemplazados y que anteriormente estaban vacios.
ndf_MeanWh.mean() 
```

```python
#En esta etapa vamos a extraer la media de los valores de cada columna de nuestro DataSet
#PAra esto primero creamos un arreglo vacio el cual contendrá nuestros nuevos valores.
Median = []
for n in name:
  Median.append(ndf_MeanWh[n].median()) #using Append to save these values in a list

Median
```

```python
#PAra poder visualizar la información sin alterarla, copiamos los elementos del DataSet y visualizamos su información
ndf_Median=df2.copy()
ndf_Median
```

```python
#En esta etapa sustituiremos todos los valores vacios del DataSet con valores promedio de la misma columna
name = ndf_Median.columns

#Para la sutitución nos apoyaremos de un ciclo for, para recorrer todas las columnas
for n in range(0,len(name)):
  j=name[n]
  ndf_Median[j].fillna(value = Median[n], inplace = True)

  
ndf_Median
```

```python
#Aplicamos la función mean, para obtener la media de los datos NA, que previamente fueron cambiados por la mediana.
ndf_Median.mean() 
```

```python
#Creamos una copia de nuestro dataframe, para no modificar los cambios anteriormente realizados.
ndf_Mode=df2.copy()
ndf_Mode
```

```python
#En esta etapa obtenedemos la moda de cada valor de nuestro DataSet
name = ndf_Mode.columns
print(name)

#Creamos un nuevo arreglo vacio el cual contendrá los valores de la moda que se obtendrá
Mode = []
for n in name:
  Mode.append(ndf_Mode[n].mode()) #using Append to save these values in a list

print(Mode)
```

```python
#Ahora se proceden a reemplazar los valores vacios del data set con los valores de la moda
name = ndf_Mode.columns

#El siguiente ciclo nos permite recorrer el arreglo para ir reemplazando los valores
for n in range(0,len(name)):
  j=name[n]
  ndf_Mode[j].fillna(value = Mode[n], 
                     inplace = True) 
ndf_Mode
```

```python
ndf_Mode.mean() #Sacar la media de los datos NA cambiados por moda.
```

```python
#Habiendo aplicado los procedimientos de media, mediana y moda, a continuación se procede a verificar si se tiene aun valores vacios o imcompletos en el DataSet
print("Media values NA? ",ndf_MeanWh.isnull().values.any(),'\n',
      "Median values NA? ",ndf_Median.isnull().values.any(),'\n',
      "Mode values NA? ",ndf_Mode.isnull().values.any(),'\n'
      )
```

Conclusión:

Podemos apreciar que reemplazando los valores vacíos NA con la media y mediana podemos tener un data set completo, a diferencia de usar la moda. Obviamente el uso de los 3 puede cambiar dependiendo del análisis estadístico que queramos aplicar.


##Conclusión

Como veremos en las preguntas de la seccion 3, veremos que en los 3 casos tenemos variaciones en los datos, por ende creemos que la mejor opcion de limpieza es descartar los NA que reemplazarlos con una variable estadistica basada sobre los mismos datos.

```python
#En esta etapa final, desplegamos lel contenido de nuestro DataSet para poder observar y comparar los valores resultantes de la aplicación y reemplazo por valores de la media, mediana y moda
print('Original: \n', df2.describe(), '\n\n',
      'Estrategia de eliminar datos: \n', ndf_dropna.describe(), '\n\n',
      'Estrategia de cambiar datos NA por la Media: \n', ndf_MeanWh.describe(), '\n\n',
      'Estrategia de cambiar datos NA por la Mediana: \n', ndf_Median.describe(), '\n\n',
      'Estrategia de cambiar datos NA por la Moda: \n', ndf_Mode.describe()) 
```

##¿Cuándo es una mediana mejor en comparación con la media?

Le Media es muy simple pero peligrosa, a pesar de ser de las soluciones más usadas como en la documentación nos muestra. Tiene grandes problemas al momento de querer utilizarla, como:

1. La imputación de mean no preserva las relaciones entre variables. Esto quiere decir que, si realmente tenemos información perdida, pudiera ser que reemplazara no indique la relación correcta entre las variables.
2. La Media nos lleva a errores estándar subestimados. Esto quiere decir que, al usar una imputación de media, podemos tener errores estándar muy bajos y por ende también los p-values.

Una conclusión adicional es que, si tenemos valores con outliers bastante marcados, la mediana obtiene mucho valor ya que esta es bastante inmune a ellos y por ende tendría una cierta ventaja sobre la media.
